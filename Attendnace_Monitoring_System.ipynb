```python
import os

# Define the main folder path
main_folder = "/content/drive/MyDrive/ULNN/Images"

# Loop through each sub-folder
for folder in os.listdir(main_folder):
    folder_path = os.path.join(main_folder, folder)

    if os.path.isdir(folder_path):  # Ensure it's a folder
        files = sorted(os.listdir(folder_path))  # Sort to maintain order

        for index, file in enumerate(files, start=1):
            file_path = os.path.join(folder_path, file)
            file_extension = os.path.splitext(file)[-1]  # Get file extension (.jpg, .png, etc.)

            # Define new file name
            new_file_name = f"{folder}_{index}{file_extension}"
            new_file_path = os.path.join(folder_path, new_file_name)

            # Rename the file
            os.rename(file_path, new_file_path)
            print(f"Renamed: {file} → {new_file_name}")

print("Renaming completed successfully!")
```

## Updating CSV file with no of photos of a student


```python
import os
import pandas as pd

# Load CSV file
csv_path = "/content/drive/MyDrive/ULNN/Attendance.csv"
df = pd.read_csv(csv_path)

# Function to count images in each folder
def count_images(folder_path):
    try:
        return len([f for f in os.listdir(folder_path) if f.endswith(('.jpg', '.png', '.jpeg'))])
    except FileNotFoundError:
        return 0

# Update "No of Photos" column
df["No of Photos"] = df["Path"].apply(count_images)

# Save updated CSV
df.to_csv("/content/drive/MyDrive/ULNN/Dataset.csv", index=False)

print("CSV file updated with the number of photos per student.")
```

```python
import os
import cv2
import pandas as pd
from keras_facenet import FaceNet

# Load FaceNet model
embedder = FaceNet()

# Load CSV file
csv_path = "/content/drive/MyDrive/ULNN/Dataset.csv"
df = pd.read_csv(csv_path)

# Loop through each student folder
for index, row in df.iterrows():
    folder_path = row["Path"]  # Get folder path from CSV

    # Ensure folder exists
    if not os.path.exists(folder_path):
        print(f"❌ Folder does not exist: {folder_path}")
        continue  # Skips this row and moves to the next student

    # List all image files in the folder
    images = [f for f in os.listdir(folder_path) if f.endswith(('.jpg', '.png', '.jpeg'))]

    if not images:
        print(f"⚠️ No images found in: {folder_path}")
        continue  # Skips if no images are found

    # Process each image
    for image_name in images:
        image_path = os.path.join(folder_path, image_name)

        # Load the image
        img = cv2.imread(image_path)
        if img is None:
            print(f"❌ Could not load image: {image_path}")
            continue  # Skip this image

        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB

        # Detect faces
        faces = embedder.extract(img, threshold=0.60)  # Lowered threshold
        if not faces:
            print(f"⚠️ No faces detected in: {image_path}")
            continue  # Skip this image if no face detected
        else:
            print(f"✅ Face detected in: {image_path}")

print("✅ Debugging complete.")
```

```python
import os
import numpy as np
import pandas as pd
import cv2
import pickle
from keras_facenet import FaceNet
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier

# Load FaceNet model
embedder = FaceNet()

# Load student data
csv_path = "/content/drive/MyDrive/ULNN/Dataset.csv"
df = pd.read_csv(csv_path)

embeddings = []
labels = []

for index, row in df.iterrows():
    person_id = row["Person ID"]
    folder_path = row["Path"]

    images = [f for f in os.listdir(folder_path) if f.endswith(('.jpg', '.png', '.jpeg'))]

    for image_name in images:
        image_path = os.path.join(folder_path, image_name)

        img = cv2.imread(image_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        faces = embedder.extract(img, threshold=0.50)
        if faces:
            embeddings.append(faces[0]["embedding"])
            labels.append(person_id)

# Convert to NumPy arrays
X = np.array(embeddings)
y = np.array(labels)

# Encode labels
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

# Train KNN model
knn = KNeighborsClassifier(n_neighbors=3, metric="euclidean")
knn.fit(X, y)

# Save trained model and label encoder
with open("knn_model.pkl", "wb") as f:
    pickle.dump(knn, f)

with open("label_encoder.pkl", "wb") as f:
    pickle.dump(label_encoder, f)

print("✅ Face embeddings extracted and KNN model trained.")
```
